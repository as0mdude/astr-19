```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


#generate random numbers
random_numbers = np.random.uniform(0, 1, 1000)

#generate histogram
plt.hist(random_numbers, bins=100, edgecolor='k')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Uniform Random Numbers Histogram')

#save the figure as a PDF
plt.savefig('random_numbers_histogram.pdf', format='pdf')
```
