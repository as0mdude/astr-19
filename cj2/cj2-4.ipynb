```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#generate 1000 random numbers distributed in a binomial way
random_numbers = np.random.binomial(100, 0.5, 1000)

#create a histogram with 100 bins
plt.hist(random_numbers, bins=100, edgecolor='k')

#label your axes
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Binomial Random Numbers Histogram')

#save the figure as a PDF
plt.savefig('binomial_numbers_histogram.pdf', format='pdf')
```
